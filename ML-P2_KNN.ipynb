```python
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.datasets import fetch_openml
```

```python
X, y = fetch_openml('mnist_784', version=1, return_X_y=True)
```

```python
# (60K: Train) and (10K: Test)
X_train, X_test = X[:60000], X[60000:]
y_train, y_test = y[:60000], y[60000:]

print(type(X_train))
print()

# Displaying dataset summary
print('X_train = %s, Y_train = %s' % (X_train.shape, y_train.shape))
print('X_test = %s, Y_test = %s' % (X_test.shape, y_test.shape))
```

```python

# Building the model  with parameter: n_neighbors=3

neigh = KNeighborsClassifier(n_neighbors=3)
neigh.fit(X_train, y_train)

pred = neigh.predict(X_test)

print(neigh.predict(X_test))
print("PRED: ", pred)
```

```python
# extra step to see all the values in the prediction
import numpy as np
with np.printoptions(threshold=np.inf):
    print(pred)
print(type(pred))
```

```python
def knnAlgoAndLeaf(n_neighbors, algo, leaf):
    neigh6 = KNeighborsClassifier(n_neighbors=3, algorithm = algo, leaf_size = leaf)
    neigh6.fit(X_train, y_train)

    pred = neigh6.predict(X_test)

    print(neigh6.predict(X_test))
    print("PRED: ", pred)

    mean_Accu6 = neigh6.score(X_test, y_test, sample_weight=None)
    print("mean Accuracy ::", mean_Accu6)
    error6 = (1- mean_Accu6)*100
    print(error6)
    return mean_Accu6, error6
```

```python

meanA_leaf1, err_leaf1 =  knnAlgoAndLeaf(3, 'ball_tree',10)
print("mean:: ", meanA_leaf1," error:: ", err_leaf1)

meanA_leaf2, err_leaf2 =  knnAlgoAndLeaf(3, 'ball_tree',20)
print("mean:: ", meanA_leaf2," error:: ", err_leaf2)

meanA_leaf3, err_leaf3 =  knnAlgoAndLeaf(3, 'ball_tree',30)
print("mean:: ", meanA_leaf3," error:: ", err_leaf3)

meanA_leaf4, err_leaf4 =  knnAlgoAndLeaf(3, 'KDTree',10)
print("mean:: ", meanA_leaf4," error:: ", err_leaf4)

meanA_leaf5, err_leaf5 =  knnAlgoAndLeaf(3, 'KDTree',20)
print("mean:: ", meanA_leaf5," error:: ", err_leaf5)

meanA_leaf6, err_leaf6 =  knnAlgoAndLeaf(3, 'KDTree',30)
print("mean:: ", meanA_leaf6," error:: ", err_leaf6)
```
